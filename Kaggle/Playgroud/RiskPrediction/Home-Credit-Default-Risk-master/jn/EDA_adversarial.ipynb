```python
import utils
import pandas as pd
from matplotlib import pyplot as plt
```

```python
tr = utils.load_train()
te = utils.load_test()
```

```python
tr['data'] = 'train'
te['data'] = 'test'
```

```python
trte = pd.concat([tr, te], ignore_index=True)
```

```python
trte.groupby('data')['AMT_CREDIT-d-AMT_ANNUITY'].plot(kind='hist', bins=50, legend=True, alpha=0.8); plt.show()
trte.groupby('data')['AMT_CREDIT-d-AMT_ANNUITY'].plot(kind='kde',  legend=True, alpha=0.8); plt.show()
```

```python
trte.groupby('data')['PAYMENT_RATE'].plot(kind='hist', bins=50, legend=True, alpha=0.8); plt.show()
trte.groupby('data')['PAYMENT_RATE'].plot(kind='kde',  legend=True, alpha=0.8); plt.show()
```

```python
trte.groupby('data')['AMT_GOODS_PRICE-d-AMT_CREDIT'].plot(kind='hist', bins=50, legend=True, alpha=0.8); plt.show()
trte.groupby('data')['AMT_GOODS_PRICE-d-AMT_CREDIT'].plot(kind='kde',  legend=True, alpha=0.8); plt.show()
```

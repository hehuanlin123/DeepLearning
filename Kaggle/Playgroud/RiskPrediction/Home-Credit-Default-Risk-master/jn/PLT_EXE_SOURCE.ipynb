```python
def predictionDist(y, feature, feature_name='feature_name'):
    import matplotlib.pyplot as plt
    
    sub = y.to_frame('y')
    sub[feature_name] = feature
    y0 = sub[sub.y==0]
    y1 = sub[sub.y==1]
    
    y0[feature_name].name = 'TARGET:0'
    y0[feature_name].plot(kind='hist', legend=True, bins=50)
    #y0[feature_name].hist(bins=50)
    
    y1[feature_name].name = 'TARGET:1'
    y1[feature_name].plot(kind='hist', legend=True, bins=50)
    #y1[feature_name].hist(bins=50)
    
    plt.legend()
    plt.xlim(feature.min(), feature.max())
    plt.xlabel(f'{feature_name}')
    plt.title(f'{feature_name} Distribution', fontsize=13, alpha=0.5)
    plt.show()
```

```python
predictionDist(train.TARGET, train.EXT_SOURCE_1, 'EXT_SOURCE_1')
```

```python
predictionDist(train.TARGET, train.EXT_SOURCE_2, 'EXT_SOURCE_2')
```

```python
predictionDist(train.TARGET, train.EXT_SOURCE_3, 'EXT_SOURCE_3')
```

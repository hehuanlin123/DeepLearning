```python
import math

# 假设一些随机数据和权重
w = [2,-3,-3]
x = [-1,-2]

# 前向传播 
dot = w[0]*x[0] + w[1]*x[1] + w[2] 
f = 1.0 / (1 + math.exp(-dot)) # sigmoid函数  
print(dot)
print(f)

# 对神经元反向传播 
ddot = (1 - f) * f # 点积变量的梯度, 使用sigmoid函数求导 
dx = [w[0] * ddot,w[1] * ddot] # 回传到x 
dw = [x[0] * ddot,x[1] * ddot,1.0 * ddot] # 回传到w 
print(ddot)
print(dx)
print(dw)

# 完成！得到输入的梯度
```

```python
import math

x = 3 # 例子数值 
y = -4   

# 前向传播 
sigy = 1.0 / (1 + math.exp(-y)) #分子中的sigmoid        #(1) 
num = x + sigy                     #分子                #(2) 
sigx = 1.0 / (1 + math.exp(-x)) #分母中的sigmoid        #(3) 
xpy = x + y                                             #(4) 
xpysqr = xpy**2                                         #(5) 
den = sigx + xpysqr #分母                               #(6) 
invden = 1.0 / den                                      #(7) 
f = num * invden #搞定!                                 #(8)

#回传 f = num * invden 
dnum = invden #分子的梯度                               #(8) 
dinvden = num  #分母倒数的梯度                              #(8) 
#回传  invden = 1.0 / den  
dden = (-1.0 / (den ** 2)) * dinvden                   #(7) 
#回传 den = sigx + xpysqr 
dsigx = (1) * dden                                     #(6) 
dxpysqr = (1) * dden                                   #(6) 
#回传 xpysqr = xpy**2 
dxpy = (2 * xpy) * dxpysqr                             #(5) 
#回传 xpy = x + y 
dx = (1) * dxpy                                       #(4) 
dy = (1) * dxpy                                       #(4) 
#回传 sigx = 1.0 / (1 + math.exp(-x)) 
dx += ((1 - sigx) * sigx) * dsigx #Notice += !! See notes below  #(3) 
#回传 num = x + sigy 
dx += (1) * dnum                                     #(2) 
dsigy = (1) * dnum                                   #(2) 
#回传 sigy = 1.0 / (1 + math.exp(-y)) 
dy += ((1 - sigy) * sigy) * dsigy                    #(1)

#对前向传播时产生每个变量(sigy, num, sigx, xpy, xpysqr, den, invden)进行回传
print(sigy,"\n",num,"\n",sigx,"\n",xpy,"\n",xpysqr,"\n",den,"\n",invden)
print("====================")
print(dsigy,"\n",dnum,"\n",dsigx,"\n",dxpy,"\n",dxpysqr,"\n",dden,"\n",dinvden)
```

在不同分支的梯度要相加：如果变量x，y在前向传播的表达式中出现多次，那么进行反向传播的时候就要非常小 心，使用+=而不是=来累计这些变量的梯度（不然就会造成覆写）。这是遵循了在微积分中的多元链式法则，该法 则指出如果变量在线路中分支走向不同的部分，那么梯度在回传的时候，就应该进行累加。 

```python
import numpy as np

#前向传播
W = np.random.randn(5, 10)    # (1)
X = np.random.randn(10, 3)    # (2)
D = W.dot(X)                  # (3)

# 假设我们得到了D的梯度 
# 回传 D = W.dot(X)
dD = np.random.randn(*D.shape) # 和D一样的尺寸  #(3)
# 回传 W = np.random.randn(5, 10)
dW = dD.dot(X.T) #.T就是对矩阵进行转置         # (1)
# 回传 X = np.random.randn(10, 3)
dX = (W.T).dot(dD)                             # (2)

print(W,"\n",X,"\n",D)
print("\n")
print(dW,"\n",dX,"\n",dD)
```
